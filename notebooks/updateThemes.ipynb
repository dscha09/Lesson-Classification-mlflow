# Using EOPS

```python
df = pd.read_excel("../../data/sentences-20200413.xlsx")
print(len(df))
print(df.columns)
```

```python
eops = pd.read_excel("../../data/eops_query_cognitive_search_pcr_sectors.xlsx")
# df = df.rename(columns={"Sentence": "paragraph"})
print(len(eops))
print(eops.columns)
```

```python
for index, row in df.iterrows():
    projectNumber = row['Project Number']
    if(row['Sectors'] == "[]"):
        sectors = []
        for i, r in eops.iterrows():
            if r['PROJ_NUMBER'] == projectNumber:
                sector = eops.at[index, 'SECTOR']
                if sector == 'Water and other urban infrastructure and services':
                    sector == 'Water'
                if sector == 'Agriculture, natural resources and rural development':
                    sector == 'Agriculture and natural resources'
                if sector == 'Finance':
                    sector == 'Finance sector development'
                if sector == 'Public sector management':
                    sector == 'Governance and public sector management'
                if sector == 'Information and communication technology':
                    sector == 'Information and Communications Technology'
                sectors.append(sector)
        df.at[index, 'Sectors'] = list(set(sectors))
        print(df.at[index, 'id'])
        print(df.at[index, 'Sectors'])
df.to_excel('../../data/sentences-20200406.xlsx')
```

# Update Old PCRs

```python
df = pd.read_excel("../../data/pcrs-20200406.xlsx")
print(len(df))
print(df.columns)
```

```python
print(df["Sectors"].unique())
```

```python
manualSectors = pd.read_excel("../../data/manual sector.xlsx")
projectNumbers = manualSectors["Project Number"].tolist()
print(manualSectors.columns)
```

```python
for index, row in df.iterrows():
    projectNumber = row['Project Number']
    if projectNumber in projectNumbers:
        sectors = []
        for i, r in manualSectors.iterrows():
            if r['Project Number'] == projectNumber:
                sector = manualSectors.at[i, 'Sectors']
                if sector == 'Water and other urban infrastructure and services':
                    sector == 'Water'
                if sector == 'Agriculture, natural resources and rural development':
                    sector == 'Agriculture and natural resources'
                if sector == 'Finance':
                    sector == 'Finance sector development'
                if sector == 'Public sector management':
                    sector == 'Governance and public sector management'
                if sector == 'Information and communication technology':
                    sector == 'Information and Communications Technology'
                if "," in sector:
                    sectors = sector.split(", ")
                else:
                    sectors.append(sector)
        df.at[index, 'Sectors'] = list(set(sectors))
```

```python
df.to_excel('../../data/pcrs-20200415.xlsx')
```
