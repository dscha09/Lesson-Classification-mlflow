```python
df = pd.read_excel('../data/merged_lessons_forecasted_data_reindex_20200226.xlsx')
PCRDataFileName = '../data/pcrs_2020-02-25.xlsx'
PCRsData = pd.read_excel(PCRDataFileName)
```

```python
print(PCRsData['UNQ_MODALITY_FR_DUMP'].unique())
```

```python
df['Modalities'] = ""
modalities = []
for index, row in df.iterrows():
    if(PCRsData[PCRsData['File Names']==df.at[index, 'File Names']].index.values):
        PCRsDataIndex = PCRsData[PCRsData['File Names']==df.at[index, 'File Names']].index.values[0]
        if(PCRsData.at[PCRsDataIndex, 'UNQ_MODALITY_FR_DUMP'] == 'set()'):
            modalities.append([])
        else:
            print([PCRsData.at[PCRsDataIndex, 'UNQ_MODALITY_FR_DUMP'][2:-2]])
            modalities.append([PCRsData.at[PCRsDataIndex, 'UNQ_MODALITY_FR_DUMP'][2:-2]])
    else:
        modalities.append([])
df['Modalities'] = modalities
```
