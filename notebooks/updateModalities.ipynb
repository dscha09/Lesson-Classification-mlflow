```python
df = pd.read_excel("../../data/sentences-20200420.xlsx")
print(len(df))
print(df.columns)
```

```python
manualModalities = pd.read_excel("../../data/manual modality.xlsx")
print(len(manualModalities))
print(manualModalities.columns)
```

```python
projectNumbers = manualModalities['Project Number'].tolist()
modalities = manualModalities['Modalities'].tolist()
print(modalities)
tmp = []
for m in modalities:
    if m == 'Project':
        tmp.append(["Project loan"])
    elif m == "Program":
        tmp.append(["Program loan"])
    else:
        tmp.append([m])
modalities = tmp
print(modalities)
```

```python
for index, row in df.iterrows():
    projectNumber = row["Project Number"]
    if(projectNumber in projectNumbers):
        df.at[index, 'Modalities'] = modalities[projectNumbers.index(projectNumber)]
        print(df.at[index, 'id'])
        print(df.at[index, 'Modalities'])
```
